# Scaling
- Min Max Scaler
- Standard Scaler
- Robust Scaler 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips[['tip','total_bill']].describe()
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
sns.distplot(tips['tip'])
plt.subplot(1,2,2)
sns.distplot(tips['total_bill'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler

minmax = MinMaxScaler()
standard = StandardScaler()
robust = RobustScaler()
```

```python
x = tips[['tip','total_bill']]
x
```

> ## Min Max Scaler
Transform each variable to range 0 - 1

```python
x_minmax = minmax.fit_transform(x)
x_minmax = pd.DataFrame(x_minmax, columns=['tip','total_bill'])
x_minmax
```

```python
x_minmax.describe()
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
sns.distplot(x_minmax['tip'])
plt.subplot(1,2,2)
sns.distplot(x_minmax['total_bill'])
plt.show()
```

> ## Standard Scaler
Transform each variable to mean = 0 and std = 1

```python
x_standard = standard.fit_transform(x)
x_standard = pd.DataFrame(x_standard, columns=['tip','total_bill'])
x_standard
```

```python
x_standard.describe()
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
sns.distplot(x_standard['tip'])
plt.subplot(1,2,2)
sns.distplot(x_standard['total_bill'])
plt.show()
```

> ## Robust Scaler

```python
x_robust = robust.fit_transform(x)
x_robust = pd.DataFrame(x_robust, columns=['tip','total_bill'])
x_robust
```

```python
x_robust.describe()
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
sns.distplot(x_robust['tip'])
plt.subplot(1,2,2)
sns.distplot(x_robust['total_bill'])
plt.show()
```
