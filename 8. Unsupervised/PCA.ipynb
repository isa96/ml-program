# Principal Component Analysis (PCA)
Data : Breast Cancer     
Features : 30 features extracted from images    
Can't see the hidden structure in data easily using data visualization. If use scatter plot, dataset has 30 features, which would result 30 * 14 = 420 scatter plots.

Task :
- Scale the features using Standard Scalar
- Compute how much information extracted using two components
- Reduce the dimension from 30 to 2 using two components
- Plot the first components and the second components
    - x axis first component
    - y axis second component
    - hue : Diagnosis
    
> ## Library and Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

cancer = load_breast_cancer()
```

```python
cancer
```

> ## Scaler

so that no variables are too dominating

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(cancer.data)
```

```python
x_scaled.shape
```

```python
cancer.feature_names
```

> ## PCA

```python
pca = PCA(n_components = 30)
x_pca = pca.fit_transform(x_scaled)
```

```python
pd.DataFrame(x_pca)
```

```python
two_pca = pd.DataFrame(x_pca)[[0, 1]]
two_pca
```

```python
variance = []

for i in range(0, 30):
    variance.append(np.var(x_pca[:, i]))
```

```python
print('information extracted :', np.sum(variance[:2])/np.sum(variance)*100, '%')
```

> ## Analyze using Scatter Plot

```python
plt.figure(figsize=(10,10))
sns.scatterplot(two_pca[0],
               two_pca[1],
               hue = np.where(cancer.target, 'benign', 'malignant'))
plt.xlabel('First Component', size = 13)
plt.ylabel('Second Component', size = 13)
plt.plot()
```

```python
cancer.target_names
```

> ## Conclusion :
1. Both components represent 63.24 % from all features
2. Both components can divide 'benign' and 'malignant' well
3. There are hidden structures to be obtained
4. 'Malignant' cancer type more variable than 'Benign' cancer type, because the scatter plot is more spread out
